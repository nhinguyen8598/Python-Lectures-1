```python
# https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html
sex = pd.get_dummies(df,columns=['Sex'])
sex
```

```python
embarked_df = pd.get_dummies(sex,columns=['Embarked'])
embarked_df
```

```python
embarked_df.drop(['PassengerId','Name','Cabin','Ticket'], axis = 1,inplace =True)
```

```python
embarked_df
```

```python
# Divide two field ( train field and test field)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
pclass = pd.get_dummies(embarked_df,columns =['Pclass'])
pclass
```

```python
y = pclass['Survived']
X = pclass.drop('Survived', axis = 1)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
logmodel = LogisticRegression()
```

```python
# random_state : sá»‘ láº§n mÃ  sÃ³c bao nhiÃªu láº§n trc khi nÃ³ chia data ra, sÃ³c ngáº«u nhiÃªn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =101)
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```
