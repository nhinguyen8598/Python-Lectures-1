```python
# phân loại nhóm tuổi của data trong info_df. Tạo 1 new col trong info_df tên là Age_Group ,rules:
# < 18 tuôi : 'Kid'
# 18 <= Age < 30 : 'Teenager'
# 30 <= Age < 55 : 'Middle_Age'
# Age >= 55 : 'Old'
# info_df['Age'][i] < 18: 1 ô dữ liệu nhỏ so sánh vs 1 gtri khác 
# info_df[info_df['Age'] <18] lấy 1 tập dữ liệu lớn để so sánh vs 1 con số
age_list = []
for i in range(len(info_df)):
    if info_df['Age'][i] < 18: 
        age_list.append('Kid')
    elif info_df['Age'][i] < 30: # >=18
        age_list.append('Teenager')
    elif info_df['Age'][i] < 55: #>= 30
        age_list.append('Middle_Age')
    else: # >=55 or NaN
        age_list.append('Old') 
    
info_df['Age_Group'] = age_list
```
