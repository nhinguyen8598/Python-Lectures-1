```python
# fill na 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html
df_age_cabin = df[['PassengerId','Age','Cabin']]
```

```python
df_age_cabin 
```

```python
# fixed fill : tính gtri xác định: mean, median (50%), q1(25%), q3(75%)
# sdung backfill, forward fill thì khai báo method còn sdung fixed fill thì khai báo value
def cal_mean(l):
    s = 0
    for i in l.dropna():
        s = s + i
    avg = s/len(l)
    return round(avg,2)  #round là làm tròn
```

```python
cal_mean(df_age_cabin['Age'])
```

```python
df_age_cabin['Age'].fillna(value = cal_mean(df_age_cabin['Age']), inplace = True)
```

```python
df_age_cabin
```

```python
#cabin là dạng data ko tính toán thì forward fill or backfill
df_age_cabin.fillna(method ='ffill')
```
