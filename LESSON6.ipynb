```python
# cách 2 
df['Age'] = df['Age'].fillna(method ='ffill')
df
```

```python
# Group 'Age' col into  4 groups, create a new col 'Age_Group':
#  < 18 :"kid"
# 18 <= Age < 30 : 'teenager'
# 30 <= Age < 45 : 'middle_age'
# Age >= 55 : elderly"
# cách 1:
age_list = []
for i in range(len(df)):
    if df['Age'][i] < 18: 
        age_list.append('kid')
    elif 18 <= df['Age'][i] < 30: # >= 18
        age_list.append('teenager')
    elif 30 <= df['Age'][i] < 55: # >= 30
        age_list.append('middle_age')
    else: # >= 55 or NaN
        age_list.append('elderly') 
    
df['Age_Group'] = age_list
```

```python
df
```

```python
# cách 2 :
age_group =[]
for age in df['Age']:
    if age < 18:
        age_group.append('kid')
    elif age < 30:
        age_group.append('teenager')
    elif age < 55:
        age_group.append('middle_age')
    else:
        age_group.append('elderly')
df['Age_Group'] = age_group
```

```python
df
```

```python
# Insight 5 : which age group is more likely to survive among the two gender groups
pivot_5 = pd.pivot_table(df, index =['Sex','Age_Group'], values ='Survived', aggfunc =np.mean)
pivot_5
```

```python
# Câu hỏi pvan 
# Write a function named check_palindrome that takes in an a positive integer as input
# This function should return  True if the inputted number is a Palindrome number 
# For example : check_palindrome(12321) should return True
#               check_palindrome(1242) should return False

# ý tưởng là sdung vòng lặp while sau đó casting về string
# vd 123456 sao lấy 6 ra thì chia lấy dư %
# còn lấy phần nguyên là 12345 thì //
def check_palindrome(n):
    num = n
    reversed_num = ''
    while num > 0:
        last_num = num % 10
        reversed_num = reversed_num + str(last_num)
        num = num // 10
    if reversed_num == str(n):
        return True
    else:
        return False
```
